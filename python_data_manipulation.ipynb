## Data Visualisation with Matplotlib

```python
import pandas as pd
df = pd.read_csv('QueryResults.csv', names = ['DATE','TAG', 'POSTS'], header = 0)
df
```

```python
df.shape
```

```python
df.count()
```

```python
df[['POSTS','TAG']].groupby('TAG').sum()
```

```python
df.DATE[0]
```

```python
type(df.DATE[0])
```

####Change to Timestamp for timeseries inspection

```python
type(pd.to_datetime(df.DATE[0]))
```

```python
df.DATE = pd.to_datetime(df.DATE)
df.head()
```

#### Pivoting

```python
reshaped_df = df.pivot(index = 'DATE', columns = 'TAG', values = 'POSTS')
```

```python
reshaped_df.shape
```

```python
reshaped_df.columns
```

```python
reshaped_df.head()
```

When pivoted, NAN is filled to when there is no post

```python
reshaped_df.count()
```

```python
reshaped_df.fillna(0, inplace = True)
```

```python
reshaped_df.isna().values.any()
```

#### Visualize

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(reshaped_df.index, reshaped_df['java'])
```

```python
plt.figure(figsize = (16,10))
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Posts', fontsize = 14)
plt.ylim(0,35000)
plt.plot(reshaped_df.index, reshaped_df.java)
```

```python
plt.figure(figsize = (16,10))
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Posts', fontsize = 14)
plt.ylim(0,35000)

for column in reshaped_df.columns:
  plt.plot(reshaped_df.index, reshaped_df[column])
```

```python
plt.figure(figsize = (16,10))
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Posts', fontsize = 14)
plt.ylim(0,35000)

for column in reshaped_df.columns:
  plt.plot(reshaped_df.index, reshaped_df[column],
           linewidth = 3, label = reshaped_df[column].name) # label for legend
plt.legend(fontsize = 16)
```

#### Make Time Series smoother with Pandas

```python
roll_df = reshaped_df.rolling(window = 3).mean() # smoother with bigger window!

plt.figure(figsize = (16,10))
plt.xticks(fontsize = 14)
plt.yticks(fontsize = 14)
plt.xlabel('Date', fontsize = 14)
plt.ylabel('Number of Posts', fontsize = 14)
plt.ylim(0,35000)

for column in roll_df.columns:
  plt.plot(roll_df.index, roll_df[column],
           linewidth = 3, label = roll_df[column].name)
plt.legend(fontsize = 16)
```
